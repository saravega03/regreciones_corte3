```python
import pandas as pd
datos = pd.read_excel('/content/drive/MyDrive/Estadistica  /corte 3/datos_industrias (1).xlsx')
datos.head()
```

```python
# prompt: necesito un analisis de regresion lineal simple que la variable X sea "Empleados" y la variable Y sea "Tasa_Crecimiento" de nuestra base de datos "datos

import statsmodels.api as sm
import matplotlib.pyplot as plt

# Define las variables dependiente e independiente
X = datos['Empleados']
Y = datos['Tasa_Crecimiento']

# Agrega una constante a la variable independiente (necesario para statsmodels)
X = sm.add_constant(X)

# Crea el modelo de regresión lineal simple
modelo = sm.OLS(Y, X)

# Ajusta el modelo a los datos
resultados = modelo.fit()

# Imprime el resumen de los resultados de la regresión
print(resultados.summary())

# Grafica los datos y la línea de regresión
plt.scatter(datos['Empleados'], datos['Tasa_Crecimiento'])
plt.plot(datos['Empleados'], resultados.predict(X), color='red')
plt.xlabel('Empleados')
plt.ylabel('Tasa_Crecimiento')
plt.title('Regresión Lineal Simple: Empleados vs. Tasa_Crecimiento')
plt.show()
```
